```python
import pandas as pd
import numpy as np

def normalize(data):
    norm = np.linalg.norm(data)
    normal_array = data/norm
    return normal_array


train_data = pd.read_csv('MNIST_training.csv')
test_data = pd.read_csv('MNIST_test.csv')

#test_data = normalize(test_data)
```

```python
from sklearn.neighbors import NearestNeighbors

# add the first test data into the training data X
# let i be the index of the test data that we want to predict
K = 10
i = 0
# X is traning data, X_test is test data
X_eval = X.append(np.transpose(pd.DataFrame(X_test.iloc[i, :]))) #what is going on here?
X_eval.index = range(0, X.shape[0] + 1)

# compute distance and top K neighbors
nbrs = NearestNeighbors(n_neighbors = K + 1).fit(X_eval)
distances, indices = nbrs.kneighbors(X_eval)
```
